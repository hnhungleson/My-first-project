```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import data base
path = '/home/hung/AnacondaProjects/Hehe/simplemaps-worldcities-basic.csv'
city_data = pd.read_csv(path)
germany_cities = city_data.iloc[2642:2697]
germany_cities
```

```python
map = Basemap(projection='ortho', lat_0=48, lon_0=11)

map.drawcoastlines()
for i in range(len(germany_cities)):
    city_location = germany_cities.iloc[i, 3:1:-1].values
    longitude, latitude = map(city_location[0], city_location[1])
    map.plot(longitude, latitude, marker='D',color='m')
plt.show()
```

```python
map = Basemap(resolution='l', 
              satellite_height=3000000.,
              projection='nsper', 
              lat_0 = 30., lon_0 = -27.,
              llcrnrx=500000.,llcrnry=500000.,urcrnrx=2700000.,urcrnry=2700000.
             )

map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral',lake_color='aqua')
map.drawcoastlines()
map.plot(longitude, latitude, marker='D',color='m')

plt.show()
```

```python
map = Basemap(projection='aeqd',
              lon_0 = 11.,
              lat_0 = 48.,
              width = 10000000,
              height = 10000000)

map.drawcoastlines()

for i in range(0, 10000000, 1000000):
    map.plot(i, i, marker='o',color='k')

plt.show()
```

```python
map = Basemap(llcrnrlon=6, urcrnrlon=16, llcrnrlat=46, urcrnrlat=55, 
              resolution="i", projection="tmerc", lat_0=48, lon_0=11)
map.drawcountries()
map.drawcoastlines()
```
